```python
import random

even_count = 0
odd_count = 0

print("Random numbers:")

for _ in range(10):
    num = random.randint(1, 100)
    print(num, end=" ")

    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("\nEven count:", even_count)
print("Odd count:", odd_count)
```
