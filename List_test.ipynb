```python
numbers = [10,15,30,3,50]
count=0
for num in numbers:
    if num % 2 ==0:
        count+=1
print("Number of event counts:",count)
```

```python
numbers = [10,15,30,3,50]
count=0
for num in numbers:
    if num % 2 !=0:
        count+=1
print("Number of odd counts:",count)
```

```python
numbers = [10,15,30,3,50]
maximum=numbers[0]
for num in numbers:
    if num >maximum:
        maximum=num
print("Number of odd counts:",maximum)
```

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
duplicates = []

for num in numbers:
    if numbers.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print(duplicates)
```
