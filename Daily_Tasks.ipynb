```python
#Print numbers from 1 to 10 except 3 and 8
i=1
while i<=10:
    if i==3 or i==8:
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#Find the first number divisible by 7 between 1 and 100
i=1
while i<=100:
    if i % 7==0:
        print(i)
        break
    i=i+1
  
```

```python
#Count vowels in a string
count=0
word =input("Enter any string ")
for i in word:
    if i in 'aeiou':
        count=count+1
print(count)
    

```
