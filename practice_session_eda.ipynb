```python
#Print 1–50, skip multiples of 3, skip numbers ending with 7, stop if number becomes 40
for i in range(1, 51):
    if i == 40:
        break
    if i % 3 == 0 or i % 10 == 7:
        continue
    print(i)
```

```python
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

common = []

for i in a:
    if i in b and i not in common:
        common.append(i)

print(common)
```

```python
s = "programming"
duplicates = []

for ch in s:
    if s.count(ch) > 1 and ch not in duplicates:
        duplicates.append(ch)

print(duplicates)
```

```python
s = "Programming"
result = ""

for ch in s:
    if ch.lower() in "aeiou":
        result += "*"
    else:
        result += ch

print(result)
```
