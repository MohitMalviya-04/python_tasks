```python
#3. Count vowels and consonants in a string(sentence) except spaces.
Sent="Hello World"
vowels=""
consonants = ""
for ch in Sent.lower():
    if ch ==" ":
        continue
    elif ch in"aeio":
        vowels +=ch
    else:
        consonants +=ch
print("Vowels:" , vowels)
print("Consonants:" , consonants)

    
```

```python
#4. Check whether a string is palindrome or not.
S = "121"
rev = ""
for ch in S:
    rev = ch + rev
if S ==rev:
    print("Plaindrome")
else:
    print("Not Plaindrome")
```

```python
#6. Remove duplicate elements from a list without using set().
list1 = [12,13,12,14,15,16,17]
list2 =list(dict.fromkeys(list1))
print(list2)
```

```python
#7. Separate even and odd numbers from a list.
list1 = [12,2,4,3,1,0,5]
even  = []
odd=[]
for n in list1:
    if n%2 ==0:
        even.append(n)
    else:
        odd.append(n)
print("Even:" ,even)
print(" Odd:",odd)
```

```python
#10. Count the frequency of elements in a tuple.
tup = (1,2,3,3,4,1,2)
freq = {}
for val in tup:
    if val in freq:
        freq[val] +=1
    else:
        freq[val]=1

print(freq)
```

```python
#13. Find the key with maximum value in a dictionary.  

dic ={'A': 1, 'b': 41, 'v': 341, 'd': 3}
max_key =max(dic,key=dic.get)
print(max_key)
```

```python
#12. Remove all elements from a set one by one.
S = {1,2,3,4,5}
for elem in list(S):
    S.remove(elem)
    print("Removed ",elem)
print (S)
```
